다음은 통계학 전공 학생과 컴퓨터과학 전공(COMP) 학생의 통계학 과목에 대한 중간고사 성적과 기말시험 성적을 나타낸 자료이다
statscore.csv

```python
import pandas as pd
df = pd.read_csv('statscore.csv')
df.head()
```

1) 각 전공별로 중간고사 및 기말성적에 대한 기술통계량을 구하고, 비교, 분석하시오

```python
df.describe()
```

2) 각 전공별로 중간고사 및 기말고사에 대한 히스토그램을 구리고, 분석하시오

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
StatMid = df.loc[df.major=='STAT', 'midterm']
CompMid = df.loc[df.major=='COMP', 'midterm']
args = dict(alpha=0.5, bins=10)
plt.hist(StatMid, **args, color='b', label="stat")
plt.hist(CompMid, **args, color='r', label="comp")
plt.gca().set(title="Midterm Histogram", ylabel="Frequency")
plt.legend()
plt.show()         
```

```python
StatFinal = df.loc[df.major=='STAT', 'final']
CompFinal = df.loc[df.major=='COMP', 'final']
args = dict(alpha=0.5, bins=10)
plt.hist(StatFinal, **args, color='b', label="stat")
plt.hist(CompFinal, **args, color='r', label="comp")
plt.gca().set(title="Final Histogram", ylabel="Frequency")
plt.legend()
plt.show()    
```

3) 각 전공별로 중간고사 및 기말고사에 대한 상자그림과 줄기-잎 그림을 그리고 분석하시오

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.boxplot(x="midterm", data=df[df.major=="COMP"])
```

```python
sns.boxplot(x="final", data=df[df.major=="STAT"])
```
