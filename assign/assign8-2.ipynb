1985 USDA 여성 영양 관련 자료. 25~50세의 연령에서 랜덤 선택된 737명 여성 영양자료
변수는 다음과 같다
- Calcium(mg)
- Iron(mg)
- Protein(g)
- Vitamin A(μg)
- Vitamin C(mg)

자료 출처 : [https://online.stat.psu.edu/stat505/lesson/1/1.4](https://online.stat.psu.edu/stat505/lesson/1/1.4)

```python
import pandas as pd
df = pd.read_csv('nutreint.csv', header=None, names=["Calcium","Iron","Protein","Vitamin A", "Vitamin C"])
df.head()
```

```python
df.dropna()          # 결측값 있는 행 제거
df.isnull().sum()    # 결측치 확인
```

```python
df.describe()       # 기술통계량 
```
